```python
X = df.drop(["Unnamed: 0", "id", "name", "is_hazardous_asteroid", "is_sentry_object", "epoch_osculation"], axis = 1)
y = df[["is_hazardous_asteroid"]].astype('bool')
```

```python
# Import label encoder
from sklearn import preprocessing
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'.
y = label_encoder.fit_transform(y)
print(y)
```

```python
y = pd.DataFrame({'is_hazardous_asteroid': y})
```

```python
y
```

```python
attributes = pd.DataFrame(df.columns)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Assuming you have your feature space (X) and binary target (y) ready
# X should be a 2D array-like structure (e.g., pandas DataFrame or NumPy array)
# y should be a 1D array-like structure (e.g., pandas Series or NumPy array)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Lists to store train and test accuracies for each model
train_accuracies = []
test_accuracies = []

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train_scaled, y_train)
dt_train_predictions = dt_classifier.predict(X_train_scaled)
dt_test_predictions = dt_classifier.predict(X_test_scaled)
train_accuracies.append(accuracy_score(y_train, dt_train_predictions))
test_accuracies.append(accuracy_score(y_test, dt_test_predictions))

# Naive Bayes Classifier (Gaussian Naive Bayes for numerical data)
nb_classifier = GaussianNB()
nb_classifier.fit(X_train_scaled, y_train)
nb_train_predictions = nb_classifier.predict(X_train_scaled)
nb_test_predictions = nb_classifier.predict(X_test_scaled)
train_accuracies.append(accuracy_score(y_train, nb_train_predictions))
test_accuracies.append(accuracy_score(y_test, nb_test_predictions))

# K-Nearest Neighbors (KNN)
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train_scaled, y_train)
knn_train_predictions = knn_classifier.predict(X_train_scaled)
knn_test_predictions = knn_classifier.predict(X_test_scaled)
train_accuracies.append(accuracy_score(y_train, knn_train_predictions))
test_accuracies.append(accuracy_score(y_test, knn_test_predictions))

# Plot the accuracies
models = ['Decision Tree', 'Naive Bayes', 'KNN']
plt.figure(figsize=(10, 6))
plt.bar(models, train_accuracies, label='Train Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Train Accuracies for Different Models')
plt.legend()
plt.show()

models = ['Decision Tree', 'Naive Bayes', 'KNN']
plt.figure(figsize=(10, 6))
plt.bar(models, test_accuracies, label='Test Accuracy')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Test Accuracies for Different Models')
plt.legend()
plt.show()
```

```python
print(train_accuracies, test_accuracies)
```

```python
from sklearn.metrics import confusion_matrix
cm_dt = confusion_matrix(y_test, dt_test_predictions)
cm_nb = confusion_matrix(y_test, nb_test_predictions)
cm_knn = confusion_matrix(y_test, knn_test_predictions)
```
